```python
import pandas as pd
data=pd.read_csv("data/meta_data.tsv",encoding='utf-8',sep='\t')
data.head()
```

# 数据清洗与预处理

```python
data.info()
```

```python
data=data.dropna()
data.info()
```

```python
data
```

```python
data.index=range(len(data))
data
```

```python
f=open("data/stop_words.txt",encoding='utf-8')
stop_words=[line.strip() for line in f]
```

```python
import jieba
features_clean=[]
for i in range(len(data)):
    s=data['title'][i]+'。'+data['abstract'][i]
    s_words=jieba.cut(s)
    s_cleans=[word for word in s_words if word not in stop_words]
    s_clean=''
    for word in s_cleans:
        s_clean+=word+' '
    features_clean.append(s_clean)
data['features_clean']=features_clean
data.head()
```

```python
labels_clean=[]
for s in data['labels']:
    ss=s.split()
    labels_clean.append(ss)
data['labels_clean']=labels_clean
data.head()
```

# 文本转换为频率/编码

```python
from sklearn.feature_extraction.text import TfidfVectorizer
document=[features_all]
tfidf_model=TfidfVectorizer(token_pattern=r"(?u)\b\w+\b").fit(document)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
multilabel_binarizer=MultiLabelBinarizer()
multilabel_binarizer.fit(data['labels_clean'])
y=multilabel_binarizer.transform(data['labels_clean'])
y.shape
```

```python
X=data.loc[:,'features_clean']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
X_train_tfidf=tfidf_model.transform(X_train)
X_train_tfidf.shape
```

```python
X_test_tfidf=tfidf_model.transform(X_test)
X_test_tfidf.shape
```
